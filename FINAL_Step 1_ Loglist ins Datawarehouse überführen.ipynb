```python
def read_html_file(filename, encoding="utf-8"):
    with open(filename, "r", encoding=encoding) as f:
        text = f.read()
    return text
```

```python
def process_html(text):
    items = text.replace("\n", " ").lower().split(" ")
    items = [i for i in items if len(i) > 1 and i not in stopwords_list]
    return items
```

```python
def process_newspaper(newspaper):
    filename = newspaper["file_name"]
    encoding = newspaper["encoding"].lower()
    
    html_code = read_html_file(filename, encoding)
    
    bstext = BeautifulSoup(html_code, "html.parser").find_all("h3")
    #items = bstext.replace("\n", " ").lower().split(" ")
    #items = [i for i in items if len(i) > 4 and i not in stopwords_list]

    count = pd.Series(items).value_counts().to_frame()
    
    count.columns = ["count"]
    count["word"] = count.index
    count["paper"] = newspaper["name"]
    count["date"] = newspaper["date"]
    return bstext
```

```python
now_str = []
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

start_date = datetime(2021, 3, 3)
end_date = datetime(2022, 4, 18)

for single_date in daterange(start_date, end_date):
    now_str.append(single_date.strftime("%Y-%m-%d")) 
print(single_date.strftime("%Y-%m-%d"))
```

```python
for i in now_str[0:4]:
    collection = []
    log_file_name = os.path.join(
        STORAGE_PATH,
        f"{i}.csv",
    )
    print(log_file_name)
    log_file = pd.read_csv(log_file_name)
    log_file.apply(process_wrapper, axis=1)
    data = pd.concat(collection, axis=0)
    print("Data shape:", data.shape)
    connection = sqlite3.connect(SQL_PATH)
    data.to_sql("wordcount", connection, index=False, if_exists="append")
```

```python
def process_wrapper(newspaper):
    name = newspaper["name"]
    try:
        count = process_newspaper(newspaper)
        #print(f"[INFO] Processing {name}")
        return count
    except:
        print(f"[ERROR] Failt to process {name}")

def process_one_logfile(log_file_name=None):
    
    collection = []
    
    if not log_file_name:
        now_str = datetime.now().strftime("%Y-%m-%d")
        log_file_name = os.path.join(
            STORAGE_PATH,
            f"{now_str}.csv",
        )

    print("[INFO] Processing", log_file_name)
    log_file = pd.read_csv(log_file_name)

    for i, newspaper in log_file.iterrows():
        count = process_wrapper(newspaper)
        if type(count).__name__ == "DataFrame":
            collection.append(count)
    
    data = pd.concat(collection, axis=0)
    print("Data shape:", data.shape)
    
    connection = sqlite3.connect(SQL_PATH)
    data.to_sql("wordcount", connection, index=False, if_exists="append")
    log_file.head()

if __name__ == "__main__":
    log_files = glob(os.path.join(STORAGE_PATH, "*.csv"))
    log_files = sorted(log_files)
    for log_file_name in log_files:
        process_one_logfile(log_file_name)
```
